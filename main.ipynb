```python
def make_table(limit_user,limit_book,df):
    df_user = df.groupby("User-ID")["Book-Rating"].sum().reset_index()
    #print(df_user)
    df_user = df_user[df_user["Book-Rating"]>limit_user]["User-ID"].tolist()
    
    df_rating = df.groupby("Book-Title")["Book-Rating"].sum().reset_index()
    #print(df_rating)
    df_rating = df_rating[df_rating["Book-Rating"]>limit_book]["Book-Title"].tolist()
    #print(df_rating)
    df_new = df[df["User-ID"].isin(df_user)]
    #print(df_new)
    df_new = df_new[df_new["Book-Title"].isin(df_rating)]
    print(df_new)
    final_table = df_new.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating",fill_value = 0)
    #print(final_table)
    return final_table
    
    
```

```python
table = make_table(200,50,book_rating)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
similarity_scr = cosine_similarity(table)
```

```python
def recommend(name):
    index = np.where(table.index==name)[0][0]
    similar_lst = sorted(list(enumerate(similarity_scr[index])),key=lambda x:x[1],reverse=True)[1:6]
    names_rec = [table.index[i[0]] for i in similar_lst]
    return names_rec
```

```python
new_dic = {}
for i in range(0,8818):
    similar_lst = sorted(list(enumerate(similarity_scr[i])),key=lambda x:x[1],reverse=True)[1:6]
    names_rec = [table.index[i[0]] for i in similar_lst]
    new_dic[table.index[i]] = names_rec
```

```python
df_recommendations = pd.DataFrame([
    {"Book": key, "Recommendations": "@ ".join(value)}
    for key, value in new_dic.items()
])

# Save the DataFrame as a CSV
df_recommendations.to_csv("recommendations.csv", index=False)
```
